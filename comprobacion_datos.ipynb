```python
import numpy as np

# Primer paso: todos los IDs del 1 al 1007 (cada cliente al menos una vez)
base_ids = np.arange(1, 1008)  # 1 a 1007 inclusive

# Si hay más reservas que clientes, generamos IDs adicionales aleatorios
restantes = n_reservas - len(base_ids)

if restantes > 0:
    adicionales = np.random.randint(1, 1008, size=restantes)
    all_ids = np.concatenate([base_ids, adicionales])
else:
    all_ids = base_ids

# Mezclamos para que no queden ordenados
np.random.shuffle(all_ids)

# Asignamos al dataframe
df_reservas['id_cliente'] = all_ids[:n_reservas]
```

```python
habitaciones = pd.read_csv('data/habitaciones.csv')

# Crear diccionario: tipo -> lista de IDs
habitaciones_dict = {}
for tipo in habitaciones['tipo'].unique():
    habitaciones_dict[tipo] = list(habitaciones[habitaciones['tipo'] == tipo]['id_habitacion'])
```

```python
def asignar_tipo_habitacion(row):
    # Suites
    if row['reserved_room_type'].lower() in ['h','p','l']:
        return 'suite'
    
    # Familiares
    elif row['children'] + row['babies'] > 0:
        # Elegimos interior o exterior al azar
        return np.random.choice(['familiar_interior', 'familiar_exterior'])
    
    # Dobles
    elif row['adults'] == 2:
        return np.random.choice(['doble_interior', 'doble_exterior'])
    
    # Individuales
    elif row['adults'] == 1:
        return np.random.choice(['individual_interior', 'individual_exterior'])
    
    else:
        # Default por si hay casos raros
        return 'individual_interior'
```

```python
# Primero asignamos el tipo
df_reservas['tipo_habitacion'] = df_reservas.apply(asignar_tipo_habitacion, axis=1)

# Ahora asignamos un id_habitacion disponible de ese tipo
def asignar_id_habitacion(tipo):
    return np.random.choice(habitaciones_dict[tipo])

df_reservas['id_habitacion'] = df_reservas['tipo_habitacion'].apply(asignar_id_habitacion)
```

```python
df_reservas['estado_reserva']=df_reservas['reservation_status']
df_reservas['adultos']=df_reservas['adults']
df_reservas['ninos']=df_reservas['children']
df_reservas['bebes']=df_reservas['babies']
df_reservas['deposito']=df_reservas['agent']
```

```python
import csv

# Definir tipos y tarifas
tipos_tarifas = [
    ('individual_interior', 80),
    ('individual_exterior', 95),
    ('doble_interior', 110),
    ('doble_exterior', 130),
    ('familiar_interior', 150),
    ('familiar_exterior', 170)
]

# Nombre del archivo CSV a generar
archivo_csv = 'habitaciones.csv'

with open(archivo_csv, mode='w', newline='') as file:
    writer = csv.writer(file)
    writer.writerow(['id_habitacion', 'tipo', 'tarifa'])
    
    # IDs 100 a 989 → repartir tipos
    tipo_index = 0
    for id_hab in range(100, 990):
        tipo, tarifa = tipos_tarifas[tipo_index]
        writer.writerow([id_hab, tipo, tarifa])
        tipo_index = (tipo_index + 1) % len(tipos_tarifas)
    
    # IDs 990 a 999 → suites
    for id_hab in range(990, 1000):
        writer.writerow([id_hab, 'suite', 350])

print(f"Archivo '{archivo_csv}' generado correctamente.")
```
